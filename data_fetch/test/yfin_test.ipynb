```python
from yfinance import EquityQuery

# Define the custom query
query = EquityQuery('and', [
    EquityQuery('gt', ['percentchange', 3]),
    EquityQuery('eq', ['region', 'us'])
])

# Execute the screen
response = yf.screen(query, sortField='percentchange', sortAsc=True)
response
```
